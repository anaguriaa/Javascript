**EQUAÇÃO DA RETA**

```python
from matplotlib import markers
import matplotlib.pyplot as plt  #biblioteca de visualização
from sklearn.datasets import make_classification
import numpy as np

a= -1    #pesos da reta
b= 4
c =0.4
def plotline(a, b, c):
     #ax+by+c=0
     #y=(-a*x -c)/b
  x = np.linspace(-2, 4, 50)  #visualizar a reta com os pontos xe y no espaco
  y = (-a*x -c)/b
  plt.axvline(0, -1, 1 , color='k', linewidth=1)
  plt.axvline(0, -2, 4 , color='k', linewidth=1)
  plt.plot(x,y)
  plt.grid(True)

plotline(a,b,c)
```
