# **Box Plot**


```python
import seaborn as sns
```

```python
sel = sns.boxplot(data=dados['Quantity'],orient='v',width=0.3)
sel.figure.set_size_inches(10,5)
sel.set_title('Quantity',fontsize=20)
sel.set_ylabel('Quantity',fontsize=15)
sel
```

Box PLot com Duas Variáveis

invetsigando a variável dependente (y) segundo teterminada característica

```python
sel = sns.boxplot(y='Sales', x='Quantity', data=dados, orient='v',width=0.5)
sel.figure.set_size_inches(10,5)
sel.set_title('Sales For Quantity',fontsize=20)
sel.set_ylabel('Sales and Quantity',fontsize=15)
sel
```

```python
sns.set_palette('Accent')
sns.set_style('darkgrid')
```

```python
sel = sns.boxplot(y='Sales', x='Product ID', data=dados, orient='v',width=0.5)
sel.figure.set_size_inches(10,5)
sel.set_title('Sales and Product ID',fontsize=20)
sel.set_ylabel('Sales and Product ID',fontsize=15)
sel
```

# **Distribuição de frequencia**

```python
sel = sns.distplot(dados['Quantity'])
sel.figure.set_size_inches(10,5)
sel.set_title('Distribuição de Frequencia',fontsize=20)
sel.set_ylabel('Profit and Sales',fontsize=15)
sel
```
